```
#create a list of dictionary of 3 products and their prices
product ={
    "apple":0.99,
    "banana":0.59,
    "orange":0.79
}

```

```
#add new product
product["grape"]=1.99
```

```
#update price
product["apple"]=2.99
```

```
#display all product
for products, price in product.items():
    print(f"{products}:${price}")
```

```
#tuples(immutable,oedered)
#create a tuple of days in a week
days=("monday","tuesday","wednesday","thursday","friday","saturday","sunday")
print("weekend days:",days[5:7])
print("weekenddays:",days[:5])
print("weekend:", days[::-1])#reverse order
print("weekend",days[-2:])#last two
print("weekend",days[:-2])#last two remove
```
